c) Korzystając z instrukcji warunkowej dodaj 1 do komórek macierzy `ar2`, jeżeli wartość w odpowiadających im komórkach macierzy `ar1` jest większa od 5. Następnie wyświetl zdefiniowaną macierz.

```python
ar1 = np.array([[5,3,6,4],[1,7,4,1],[9,2,8,5]])
ar2 = np.zeros_like(ar1)

for j in range(3):
    for i in range(len(ar1[1])):
        if ar1[j,i] > 5:
            ar2[i,j] = ar2[i,j] +1 
        
ar2
```
