```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
lewe=cv2.imread('left.jpg')
prawe=cv2.imread('right.jpg')
```

```python
lewe_gray=cv2.cvtColor(lewe, cv2.COLOR_BGR2GRAY)
prawe_gray=cv2.cvtColor(prawe, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(lewe_gray, cmap='gray')
```

```python
r=lewe_gray
g=prawe_gray
b=prawe_gray
```

```python
anaglif=cv2.merge([b, g, r])
```

```python
anaglif=cv2.cvtColor(anaglif, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(anaglif)
```

```python
plt.imsave('anaglif_halysfilip.jpg', anaglif)
```
